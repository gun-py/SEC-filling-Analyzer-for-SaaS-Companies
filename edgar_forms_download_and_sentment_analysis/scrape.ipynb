```python
pat = "forms/2021/QTR1/"
save_path='QTR1/'
files=os.listdir(pat)
for fl in files:
    print("At file ",fl)
    path=pat+fl
    with open(path,encoding='mbcs') as file:
        soup = BeautifulSoup(file.read())
    x = soup.find_all('type')
    y=soup.find_all('document')
    res=''
    fnd=0
    dtr = soup.find_all('sec-document')
    for i in dtr:
        k=i.getText().index(':')
        date=i.getText()[k+2:k+2+8]
        fn_dt=date[-2:]+'-'+date[4:6]+'-'+date[0:4]

    for i in range (len(x)):
        sr=x[i].getText()[0:10]
        fn=re.findall(r'EX-[0-9]+[.][0-9]+',sr)
        if len(fn)>0:
            fnd=1
            ss=y[i].getText()
            # ss=ss.replace(',',' ')
            ss=ss.replace('Â','')
            ss=ss.replace('â—','')
            res=res+ss
    if not(fnd):
        print("Did not found any Exibit looking in 8-K")
        for i in range (len(x)):
            sr=x[i].getText()[0:10]
            if '8-K' in sr:
                fnd=1
                ss=y[i].getText()
                # ss=ss.replace(',',' ')
                ss=ss.replace('Â','')
                ss=ss.replace('â—','')
                res=res+ss
    fl=open(save_path+fl[0:fl.index('.')]+'-'+fn_dt+'.txt','w+',encoding='UTF-8')
    # res=res.encode('UTF-8','ignore')
    # res=res.decode('UTF-8','ignore')

    fl.writelines(res)
```
