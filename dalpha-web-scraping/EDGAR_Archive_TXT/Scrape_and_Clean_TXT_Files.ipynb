```python
import pandas as pd
import numpy as np
import re
import requests
from bs4 import BeautifulSoup
from sklearn.cluster import KMeans
```

```python
def brace_cleaner(raw):
    cleantext = re.sub(re.compile('{.*?}'), '', raw)
    return cleantext

def get_txt_from_archives(url):
    content = requests.get(url, headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"})
    document_content = content.text
    bs4_clean = BeautifulSoup(document_content).get_text()
    clean_texts = re.sub('\\n+','\\n', brace_cleaner(bs4_clean)).replace('/s/', '').replace('\t', '')
    list_texts = [txt for txt in list(set(clean_texts.split('\n'))) if len(txt.split(' ')) > 1]
    df = pd.DataFrame(list_texts, columns = ['text'])
    df['count'] = df['text'].map(lambda x: len(x.split(' ')))
    kmeans = KMeans(n_clusters=2, random_state=2022).fit(np.array(df['count']).reshape(-1, 1))
    df['labels'] = kmeans.labels_
    grp1 = df[df['labels'] == 1]['count'].mean()
    grp2 = df[df['labels'] == 0]['count'].mean()
    flag = [0, 1][grp1 > grp2]
    unic = df[df['labels'] == flag]['text'].str.cat(sep=' ')
    encode = unic.encode("ascii", "ignore")
    decode = encode.decode()
    out = re.sub(' +',' ', decode)
    return out, df
```
